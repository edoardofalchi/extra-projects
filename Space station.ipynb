# International Space Station current location
This notebook is aimed at locating and plotting the location of the International Space Station.  
Here's the link to the API: http://open-notify.org/Open-Notify-API/ISS-Location-Now/

---
The ISS is moving at close to 28,000 km/h so its location changes really fast!  
Where is it right now?

To find it out we are interested in the latitude and longitude returned by the API.





```python
import pandas as pd
import plotly.express as px
```

```python
url = 'http://api.open-notify.org/iss-now.json'
```

```python
df = pd.read_json(url)
```

```python
df
```

```python
df['latitude']= df.loc['latitude', 'iss_position']
df['longitude']= df.loc['longitude','iss_position']
df.reset_index(inplace=True)
```

```python
df
```

```python
df = df.drop(['index','message'], axis=1)
```

```python
df
```

```python
fig = px.scatter_geo(df, lat='latitude',
                     lon='longitude')
fig.show()
```

To update its location just re-run the `url` variable command.
