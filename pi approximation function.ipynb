I want to estimate $\pi$, given that having random (0,1)

$$\frac{\pi}{4}=\frac{\text{count}}{\text{tot}}$$
$$\pi=\frac{4*count}{tot}$$

```python
def pi_approx(iter):
  count=0
  for i in range(iter+1):
    x=np.random.uniform()
    y=np.random.uniform()
    dist=(x**2+y**2)**0.5
    if dist<1:
      count+=1
  pi=(4*count/iter)
  return pi

print(pi_approx(1000000))
```
